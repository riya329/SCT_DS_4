```python
missing_values = df.isnull().sum()
print(missing_values[missing_values >0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# convert start_time to datetime format
df['Start_Time'] = pd.to_datetime(df['Start_Time'],format='mixed' , errors='coerce')

# extract hour of the day
df['Hour'] = df['Start_Time'].dt.hour

#plot accident frequency by hour
plt.figure(figsize=(10,5))
sns.countplot(x=df['Hour'] , palette="coolwarm")
plt.xlabel("Hour of the Day")
plt.ylabel("Number of Accidents")
plt.title("Accidents by Hour of the Day")
plt.show()
```

```python
# top 10 weather conditions
plt.figure(figsize=(10,5))
sns.countplot(y=df['Weather_Condition'] , order=df['Weather_Condition'].value_counts().index[:10] ,palette="magma")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Condidtion")
plt.title("Top 10 weather Condidtion in Accidents")
plt.show()
```
